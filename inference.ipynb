# IndoBERT Berita Classifier - Proses Inference

Memuat model yang sudah di-train dari `./output/saved_model` dan menjalankan prediksi klasifikasi berita.

```python
from transformers import pipeline

SAVE_DIR = "./output/saved_model"

classifier = pipeline(
    "text-classification",
    model=SAVE_DIR,
    tokenizer=SAVE_DIR,
    top_k=5,
    device="cpu",
)

print("Model loaded from:", SAVE_DIR)
```

```python
sample_texts = [
    "Presiden Prabowo meresmikan proyek infrastruktur strategis nasional di Papua",
    "IHSG ditutup menguat 1,5 persen pada perdagangan Senin",
    "Apple meluncurkan iPhone 18 dengan chip A20 Bionic terbaru",
    "Timnas Indonesia lolos ke babak semifinal Piala AFF 2026",
    "Film Dilan 3 meraih penonton terbanyak sepanjang 2026",
]

print("Prediksi Model")
print("=" * 60)
for text in sample_texts:
    result = classifier(text)
    top = result[0][0]
    print(f"\n  Teks: \"{text[:70]}...\"")
    print(f"  → Prediksi: {top['label']} ({top['score']:.2%})")
```

```python
text = "MK Tolak Gugatan terhadap Peran Suami Wajib Nafkahi Istri di UU Perkawinan"

result = classifier(text)
print(f"Teks: \"{text}\"\n")
for pred in result[0]:
    bar = "█" * int(pred['score'] * 40)
    print(f"  {pred['label']:>12}: {pred['score']:.4f} {bar}")
```
