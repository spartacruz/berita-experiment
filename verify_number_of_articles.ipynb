```python
import json
import sys

def get_array_length(json_path):
    with open(json_path, "r") as f:
        data = json.load(f)

    if isinstance(data, list):
        return len(data)
    else:
        raise ValueError("JSON root is not an array (list).")
```

```python
print("Politik: ", get_array_length('./sources/1_news-politik/3_politik_cnn_articles.json'))
print("Ekonomi: ", get_array_length('./sources/2_news-ekonomi/3_ekonomi_cnn_articles.json'))
print("Teknologi: ", get_array_length('./sources/3_news-teknologi/3_teknologi_cnn_articles.json'))
print("Olahraga: ", get_array_length('./sources/4_news-olahraga/3_olahraga_cnn_articles.json'))
print("Hiburan: ", get_array_length('./sources/5_news-hiburan/3_hiburan_cnn_articles.json'))
```
