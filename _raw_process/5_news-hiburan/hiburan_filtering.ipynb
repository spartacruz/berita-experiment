```python
import json

with open("5_hiburan_cnn_index_results.json", "r", encoding="utf-8") as f:
    data = json.load(f)
    
# print(data)
```

```python
# Extract all tags, using a set to automatically remove duplicates
unique_tags = set(item.get('tag') for item in data if item.get('tag'))
# Print the unique tags
for tag in unique_tags:
    print(tag)
```

```python
exclude_tags = {
    "GALERI DRAMA KOREA",
    "LAPORAN INTERAKTIF",
    "LPS FINANCIAL FESTIVAL",
    "VIDEO",
    "THE BIG IDEA FORUM",
}
# Filter out any items where the tag is in the exclude_tags set
filtered_data = [item for item in data if item.get('tag') not in exclude_tags]
print(f"Original count: {len(data)}")
print(f"Filtered count: {len(filtered_data)}")
```

```python
data = filtered_data
```

```python
# Title begins with
prefixes = ("FOTO:", "VIDEO:", "INFOGRAFIS:")
# Filter OUT any items where the title begins with any of the prefixes
filtered_data = [
    item for item in data 
    if not item.get('title', '').startswith(prefixes)
]
print(f"Original count: {len(data)}")
print(f"Filtered count: {len(filtered_data)}")
```

```python
data = filtered_data
```

```python
output_file = "5_hiburan_cnn_index_results_filtered.json"

with open(output_file, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)
print(f"Successfully saved {len(data)} items to {output_file}")
```
